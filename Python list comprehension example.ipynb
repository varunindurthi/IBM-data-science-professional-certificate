### This notebook showcases how to efficiently loop over multiple python lists with the help of list comprehension and zip function.

```python
a = [1,2,3]
b = [5,6,7]
c = [8,9,10]
d = [11,12,13]

print([p+q+r+s for p,q,r,s in zip(a,b,c,d)])
```
